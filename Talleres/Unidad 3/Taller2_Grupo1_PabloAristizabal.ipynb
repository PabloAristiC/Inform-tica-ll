# Conteo De CÃ©lulas

```python
import cv2
img = cv2.imread(r"Imagen.jpg", cv2.IMREAD_COLOR)
areaMinima=500
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh=cv2.threshold(gris,80,255,
                     cv2.THRESH_BINARY)
contours, _=cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
contoursFiltered=[c for c in contours if cv2.contourArea(c)>areaMinima]
imagenCountours=img.copy()
cv2.drawContours(imagenCountours, contoursFiltered,-1,(0,255,0),8)
imagen_texto=cv2.putText(imagenCountours,
                         f'Numero de celulas: {len(contoursFiltered)}',
                         (500,700),
                         cv2.FONT_HERSHEY_SIMPLEX,
                         1,
                         (255,255,255),
                         3)
cv2.imshow('Imagen con contornos', imagen_texto)
# Esperar a que se presione una tecla y cerrar la ventana
cv2.waitKey(0)
cv2.destroyAllWindows()
```
