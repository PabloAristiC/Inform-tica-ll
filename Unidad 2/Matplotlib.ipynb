```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2*np.pi,200)
y=np.sin(x)
z=np.cos(x)
fig,ax=plt.subplots(figsize=(10,8))
ax.plot(x,y,label='Grafia Seno(x)',color='gold',linewidth=(3),linestyle='--') #':','--','-','-.'
ax.plot(x,z,'d',label='Grafia Coseno(x)',color='silver',linewidth=(6)) # 's','v','o','d'
ax.set_xlabel('EJE x')
ax.set_ylabel('EJE y') #Nombres de los ejes
ax.set_title('Gráfica de Seno(X) y Coseno(x)') #Título
ax.legend() #Identificaión de linea
ax.text(3.5,0.5,'HOLAAAA') #Texto
ax.grid() #Cuadrícula
plt.show()
```

```python
x=np.linspace(0,2*np.pi,500000)
y=np.sin(x)
z=np.cos(x)
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(10,10))  #Se imprimen 2 figuras
ax1.plot(x,y,label='Grafia Seno')
ax2.plot(x,z,label='Grafia Coseno')
ax1.set_xlabel('EJE x')
ax1.set_ylabel('EJE y')
ax1.set_title('Gráfica de Seno(X)')
ax1.legend()
a=np.max(y)
ax1.annotate('Máximo local',xy=(1.5,a),xytext=(3.5,0.5),   #Construcción de la flecha
            arrowprops=dict(facecolor='black',shrink=0.05))
ax.set_ylim(-2,2)
ax2.set_xlabel('EJE x')
ax2.set_ylabel('EJE y')
ax2.set_title('Gráfica de Coseno(x)')
ax2.legend()
plt.show()
```
