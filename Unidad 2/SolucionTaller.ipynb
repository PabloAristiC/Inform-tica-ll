```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import pandas as pd
```

```python
intervalo=np.linspace(0,5,1000)
yn1=np.sin(np.pi*2*intervalo)
yn2=np.cos(np.pi*2*intervalo)
fig,ax=plt.subplots(figsize=(10,8))
ax.plot(intervalo,yn1,label='Grafia Seno(x)',color='gold',linewidth=(3)) #':','--','-','-.'
ax.plot(intervalo,yn2,label='Grafia Coseno(x)',color='silver',linewidth=(6)) # 's','v','o','d'
ax.set_xlabel('EJE x')
ax.set_ylabel('EJE y') #Nombres de los ejes
ax.set_title('Grfica de Seno(X) y Coseno(x)') #Ttulo
ax.set_ylim(-5,5)
ax.set_xlim(0,5)
ax.legend() #Identificain de linea
# ax.text(3.5,0.5,'HOLAAAA') #Texto
ax.grid() #Cuadrcula
plt.show()
```

```python
sn=yn1+yn2
tn=yn1*yn2
fig,ax=plt.subplots(figsize=(10,8))
ax.plot(intervalo,sn,label='yn1 + yn2',color='gold',linewidth=(3),linestyle='--') #':','--','-','-.'
ax.plot(intervalo,tn,label='yn1 x yn2',color='silver',linewidth=(6), linestyle=':') # 's','v','o','d'
ax.set_xlabel('EJE x')
ax.set_ylabel('EJE y') #Nombres de los ejes
ax.set_title('Grfica de sn y tn') #Ttulo
ax.legend() #Identificain de linea
# ax.text(3.5,0.5,'HOLAAAA') #Texto
ax.grid() #Cuadrcula
plt.show()
```
