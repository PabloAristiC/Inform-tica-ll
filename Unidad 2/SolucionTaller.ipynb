```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
intervalo=np.linspace(0,100,101)
yn1=np.sin(np.pi*0.12*intervalo)
yn2=np.cos(np.pi*0.03*intervalo)
fig,ax=plt.subplots(figsize=(10,8))
ax.plot(intervalo,yn1,label='Grafia Seno(x)',color='gold',linewidth=(3)) #':','--','-','-.'
ax.plot(intervalo,yn2,label='Grafia Coseno(x)',color='silver',linewidth=(6)) # 's','v','o','d'
ax.set_xlabel('EJE x')
ax.set_ylabel('EJE y') #Nombres de los ejes
ax.set_title('Gráfica de Seno(X) y Coseno(x)') #Título
ax.legend() #Identificaión de linea
# ax.text(3.5,0.5,'HOLAAAA') #Texto
ax.grid() #Cuadrícula
plt.show()
```

```python
sn=yn1+yn2
tn=yn1*yn2
fig,ax=plt.subplots(figsize=(10,8))
ax.plot(intervalo,sn,label='yn1 + yn2',color='gold',linewidth=(3),linestyle='--') #':','--','-','-.'
ax.plot(intervalo,tn,label='yn1 x yn2',color='silver',linewidth=(6), linestyle=':') # 's','v','o','d'
ax.set_xlabel('EJE x')
ax.set_ylabel('EJE y') #Nombres de los ejes
ax.set_title('Gráfica de sn y tn') #Título
ax.legend() #Identificaión de linea
# ax.text(3.5,0.5,'HOLAAAA') #Texto
ax.grid() #Cuadrícula
plt.show()
```
