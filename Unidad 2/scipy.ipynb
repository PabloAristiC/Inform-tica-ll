### Transformada de Fourier

```python
# importing the fft and inverse fft functions from fftpackage  
from scipy.fftpack import fft  
#Importing numpy  
import numpy as np  
#create an array with random n numbers  
x = np.random.rand(20)  
z=np.random.randn(20)
#Applying the fft function  
y = fft(x)  
print(x)
print(y) 
# print(x.shape)
# print(x.ndim)
```

```python
import numpy as np 
from scipy.fftpack import fft 
import matplotlib.pyplot as plt
x = np.random.rand(20)  
z=np.random.randn(20)
y = fft(x)  
print(x)
print(y)
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,10))  #Se imprimen 2 figuras
ax1.plot(x,z,label='Señal sin transformar',color="blue",linewidth=(3),linestyle='--')   #Grafica de
ax2.plot(y,z,label='Señal Transformada',color='gold',linewidth=(3),linestyle=':')
ax1.set_xlabel('Tiempo')
ax1.set_ylabel('Amplitud')
ax1.set_title('Señal Sin Transformar')
ax1.legend()
ax2.set_xlabel('frecuencia')
ax2.set_ylabel('Amplitud')
ax2.set_title('Señal Transformada')
ax2.legend()
plt.show()
```

### Interpolación

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

datos_x = np.arange(0, 0.1, 0.005)
A, k, theta = 10.0, 33.3, np.pi/5.0
y_real = A*np.sin(2*np.pi*k*datos_x + theta)
datos_y = y_real + 2*np.random.randn(len(datos_x))

interpolador_lineal = interp1d(datos_x, datos_y)


x_inter = np.linspace(0.012, 0.09, 500)
y_inter_l = interpolador_lineal(x_inter)


plt.plot(datos_x, datos_y, 'ok', label="Datos")
plt.plot(x_inter, y_inter_l, 'r', label="Interpolación lineal")
plt.title("Interpolación")
plt.xlim(-0.003, 0.102)
plt.ylim(-14, 15)
plt.legend()
```
